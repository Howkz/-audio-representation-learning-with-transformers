# Working with audio data for ML and Transformers

Audio data presents unique challenges for machine learning because it originates as a continuous-time physical signal that must be converted into a structured numerical representation. Unlike text, which is naturally discrete, or images, which are organized on a fixed grid, audio is sampled at high temporal resolution, often producing very long sequences. This makes representation and preprocessing crucial design steps. Practical systems must decide on a sampling rate, normalize amplitudes, transform the waveform into time–frequency features such as log-Mel spectrograms, and handle variable-length sequences through padding or cropping. These choices strongly influence model efficiency, stability, and generalization performance.

In modern machine learning pipelines, audio models rarely operate directly on raw waveforms; instead, they learn from structured representations that expose frequency content and reduce sequence length. Transformer-based architectures, for example, require converting spectrogram frames or patches into tokens and managing attention complexity under memory constraints. Furthermore, audio learning often benefits from self-supervised pretraining, where models learn general acoustic patterns before being fine-tuned on labeled tasks such as speech recognition. Building effective audio systems therefore requires not only architectural design, but also careful engineering of preprocessing, batching, evaluation, and reproducibility.

## From CNNs to self-supervised pre-trained models
For many years, audio machine learning systems were dominated by convolutional neural networks (CNNs). Convolutions provided a strong inductive bias for local structure in spectrograms, effectively capturing short-term temporal patterns and local frequency correlations. CNN-based acoustic models were computationally efficient, scalable, and well-suited for supervised training on labeled datasets. However, their receptive field grows gradually with depth, and modeling long-range dependencies (e.g., across entire utterances) requires stacking many layers or introducing additional recurrent components. As a result, while CNNs perform well for local pattern extraction, they are less flexible for capturing global contextual structure without architectural complexity.

The transition toward pretrained Transformer-based models, such as Whisper, reflects a broader shift in machine learning toward large-scale self-supervised learning and sequence modeling. Transformers remove the locality constraint of convolutions and instead rely on self-attention to model relationships across an entire sequence. This enables direct modeling of long-range temporal dependencies and better integration of acoustic context. Combined with large-scale pretraining on diverse multilingual datasets, models like Whisper learn robust, transferable audio representations that generalize across accents, domains, and noise conditions. This evolution mirrors similar transitions in NLP and vision: from task-specific CNN architectures trained from scratch to large pretrained Transformer encoders that serve as reusable foundations for downstream tasks.

## 1) Audio as a discrete signal

An audio signal is a continuous-time physical signal representing variations of air pressure perceived as sound. When processed by a computer, it is converted into a discrete-time digital signal through sampling.
A sound wave is continuous, but digital devices store a finite array of samples. Sampling measures the signal at fixed time steps. The sampling rate (in Hz) is the number of samples per second. For example, with 16 kHz you store 16 000 samples per second.

### Sampling
Let $x(t)$ be a continuous audio signal. Sampling at frequency $f_s$ produces a discrete signal:

$$
 x[n] = x(n / f_s)
$$

where:
- $f_s$ is the sampling rate (e.g., 16 kHz),
- $n$ is the sample index.

**Nyquist–Shannon theorem**:
To avoid aliasing, the sampling rate must satisfy:

$$
 f_s \ge 2 f_{max}
$$

where $f_{max}$ is the highest frequency in the signal.

**Common sampling rates**:
- 8 kHz: telephony
- 16 kHz: speech (ASR, TTS)
- 44.1 kHz: music (CD quality)

A key consequence is the Nyquist limit: the highest frequency you can represent is half the sampling rate. For 16 kHz, that is 8 kHz, which is sufficient for most speech content.

```python
def show_sampling_effects(freq_hz=1000, duration_s=0.02, sr_list=(7900, 16000)):
    t_cont = np.linspace(0, duration_s, int(200000*duration_s), endpoint=False)  # "continuous" grid for visualization
    x_cont = np.sin(2*np.pi*freq_hz*t_cont)

    plt.figure(figsize=(12,4))
    plt.plot(t_cont*1000, x_cont, label="continuous (viz)", alpha=0.6)

    for sr in sr_list:
        t = np.arange(int(sr*duration_s)) / sr
        x = np.sin(2*np.pi*freq_hz*t)
        plt.scatter(t*1000, x, s=18, label=f"sampled @ {sr/1000:.1f} kHz")

    plt.title(f"Sampling a {freq_hz} Hz sine wave for {duration_s*1000:.1f} ms")
    plt.xlabel("time (ms)")
    plt.ylabel("amplitude")
    plt.legend()
    plt.grid(True)
    plt.show()

show_sampling_effects(freq_hz=1000)
```

## 2) Load and inspect a real audio dataset

We will use a small dataset from the Hugging Face Hub so that experiments remain lightweight.

We load a dataset containing short audio clips. We will inspect:
the waveform array and its length, the sampling rate, the label (if any), the duration in seconds, and the dynamic range.

You must ensure that your pipeline uses a *consistent* sampling rate. If you fine-tune or compare models, mismatched sampling rates make learning harder and evaluation misleading.


```python
def decode_audio(audio_dict, target_sr=16000):
    """
    Decode audio without torchaudio / torchcodec / soundfile.

    Supported:
    - WAV via scipy
    - FLAC / OPUS / others via ffmpeg subprocess (UNIX)

    Requires:
    - ffmpeg available in PATH
    """
    # Try WAV decoding with scipy
    try:
        if audio_dict.get("bytes") is not None:
            data = io.BytesIO(audio_dict["bytes"])
            sr, wav = wavfile.read(data)
        elif audio_dict.get("path") is not None and os.path.exists(audio_dict["path"]):
            sr, wav = wavfile.read(audio_dict["path"])
        else:
            raise ValueError("No WAV source")
    except Exception:
        # Fallback: ffmpeg decode to raw PCM
        if audio_dict.get("bytes") is not None:
            # Write bytes to temporary file
            import tempfile
            with tempfile.NamedTemporaryFile(suffix=".audio") as f:
                f.write(audio_dict["bytes"])
                f.flush()
                path = f.name
        elif audio_dict.get("path") is not None:
            path = audio_dict["path"]
        else:
            raise RuntimeError("No audio source available")

        cmd = [
            "ffmpeg",
            "-i", path,
            "-f", "f32le",
            "-ac", "1",
            "-ar", str(target_sr),
            "pipe:1"
        ]

        proc = subprocess.run(
            cmd,
            stdout=subprocess.PIPE,
            stderr=subprocess.DEVNULL,
            check=True
        )

        wav = np.frombuffer(proc.stdout, dtype=np.float32)
        sr = target_sr

    # Normalize / mono safety
    if wav.ndim > 1:
        wav = wav.mean(axis=1)

    # Final resampling safety (usually not needed)
    if sr != target_sr:
        n = int(len(wav) * target_sr / sr)
        wav = scipy.signal.resample(wav, n).astype(np.float32)
        sr = target_sr

    return wav.astype(np.float32), int(sr)
```

```python
# Small and convenient: use a subset of a common audio dataset.
# IMPORTANT: we avoid HF automatic decoding (which may require torchcodec) and decode manually.
# We also avoid soundfile to keep dependencies minimal.

ds = load_dataset("superb", "ks", split="train[:1%]")  # keyword spotting subset

# Disable automatic decoding (avoids torchcodec backend issues)
ds = ds.cast_column("audio", Audio(decode=False))

print(ds)

ex = ds[0]
print(ex.keys())
print("label:", ex.get("label", None))

audio_meta = ex["audio"]
print("audio keys:", audio_meta.keys())
print("path:", audio_meta.get("path", None))
print("bytes present:", audio_meta.get("bytes") is not None)

# Decode one example to establish the global waveform used throughout the notebook
wave, sr = decode_audio(audio_meta, target_sr=16000)

print("decoded sampling_rate:", sr)
print("num_samples:", len(wave))
print("duration_s:", len(wave) / sr)
print("min/max:", float(np.min(wave)), float(np.max(wave)))
```

## Common audio loading failures

When working with audio datasets, failures often occur at the decoding stage, not at model training time. The most common issues are missing decoding backends, paths that are not resolvable, inconsistent sampling rates, and variable lengths that break batching.

In this notebook we intentionally disable automatic decoding in the Hugging Face `datasets` library to avoid dependency problems. We decode manually from either raw bytes or a valid local path. If decoding fails, print the `audio` dictionary to see which keys are available, and check whether it contains `bytes` or only a relative filename.

- If you see an error about `torchcodec`, it usually means something tried to decode the audio automatically. The fix is to cast the column with `Audio(decode=False)` and decode manually.

- If you see a `FileNotFoundError` with a filename like `doing_the_dishes.wav`, it usually means the `path` is relative and not resolvable. In that case, decoding from `bytes` is the robust approach, or you must ensure the dataset is fully cached locally.

- If your audio decodes but sounds "too fast" or "too slow", this is a sampling rate mismatch. Always resample to a target sampling rate (16 kHz is a good default for speech and many general audio tasks).

Finally, if your training dataloader crashes due to variable-length tensors, it means you must crop or pad audio to a fixed length (or implement bucketing). We implement a simple crop/pad collate function later in this notebook.


## 3) Visualizing the waveform

Audio datasets are composed of discrete-time signals obtained by sampling a continuous acoustic waveform. Each audio example is represented as a one-dimensional array $x \in \mathbb{R}^{T}$, where $T$ is the number of samples.

The temporal duration of a signal is determined by the sampling rate $f_s$ through the relation $T / f_s$. As a consequence, two signals with the same number of samples may correspond to different physical durations if their sampling rates differ. Inspecting waveform length, duration, and amplitude statistics is therefore a fundamental sanity check to understand the temporal scale of the data, the variability of examples, and potential issues such as excessive silence or clipping.

A crucial requirement in any audio learning pipeline is the use of a consistent sampling rate across all examples. Feature extraction methods (e.g., spectrograms or Mel filterbanks) and downstream neural models implicitly assume a fixed time–frequency resolution defined by $f_s$.

If audio clips with different sampling rates are mixed, the same acoustic event is mapped to different discrete representations, introducing spurious variability that degrades learning and invalidates comparisons between models. For this reason, all audio signals are systematically resampled to a common target rate $f_s^*$ before batching, feature extraction, and training, ensuring that the model operates on a coherent and physically meaningful representation of sound.

### Representation and why raw waveforms are difficult for Transformers
A raw audio waveform is a high-resolution temporal signal represented as a sequence
$x \in \mathbb{R}^{T}$, where $T$ is proportional to the sampling rate and the signal duration. Even short audio clips quickly lead to very large sequence lengths. For example, a 1-second signal at 16 kHz already yields $T = 16{,}000$ tokens. Since self-attention computes pairwise interactions between all time steps, the computational and memory complexity of a Transformer layer scales as

$$
 \mathcal{O}(T^2)
$$

which becomes prohibitive for raw waveforms, both in training time and GPU memory usage.
Beyond sequence length, raw waveforms exhibit strong local correlations: consecutive samples are highly dependent due to the physical continuity of sound. However, standard self-attention treats all token pairs symmetrically and does not impose any locality bias. As a result, the model must learn from scratch that nearby samples are more informative than distant ones, which is statistically inefficient and slows down convergence compared to architectures with built-in locality (e.g., convolutions).

Finally, most semantic information in audio is expressed in the frequency domain (pitch, timbre, formants, harmonics), not directly in the raw amplitude values. Transformers operating on waveforms lack an inductive bias for frequency structure and periodicity; they must implicitly infer these properties through many layers and attention heads. Time-frequency representations such as spectrograms or log-Mel features explicitly expose this structure, drastically reducing sequence length while aligning the input representation with both human perception and the statistical regularities of audio signals.

## 4) Time–Frequency Analysis

Raw audio waveforms encode sound as a function of time, but most meaningful acoustic information (pitch, timbre, formants, harmonics) is expressed in the frequency domain. A standard Fourier Transform converts a signal entirely into frequencies, but it loses all temporal localization. Audio signals are non-stationary: their frequency content changes over time. Time–frequency analysis addresses this by jointly representing when and at which frequency energy occurs.

### Fourier Transform
The discrete Fourier Transform (DFT) of a signal $x[n]$ is defined as:
$$
X(k)=\sum^{T-1}_{n=0}​x[n]e^{−j2\pi kn/T}
$$

where:
- $k \in {0, \dots, T-1}$ indexes frequency bins
- $X(k) \in \mathbb{C}$ is the complex-valued Fourier coefficient at frequency index $k$
- $j$ denotes the imaginary unit ($j^2 = -1$)
- the exponential term $e^{-j2\pi kn/T}$ corresponds to a complex sinusoid of normalized frequency $k/T$

This representation assumes the signal is stationary over its entire duration, which is rarely true for speech or environmental sounds. As a result, we cannot localize events such as phoneme boundaries, transients, or rhythm changes in time.
To overcome this limitation, we apply the Fourier Transform locally in time.

A spectrum shows which frequencies are present in a short segment of the signal. It is computed via a DFT. The DFT output is complex-valued; many ML pipelines keep magnitudes (or power) and discard phase.

Below we compute a spectrum for a short window of samples and display it in dB.

```python
def plot_spectrum(x, sr, n_fft=2048, start_sample=0, title="Amplitude spectrum (dB)"):
    x = np.asarray(x)
    seg = x[start_sample:start_sample+n_fft]
    if len(seg) < n_fft:
        seg = np.pad(seg, (0, n_fft-len(seg)))
    window = np.hanning(n_fft)
    seg = seg * window
    dft = np.fft.rfft(seg)
    amp = np.abs(dft) + 1e-12
    amp_db = 20*np.log10(amp / np.max(amp))
    freq = np.fft.rfftfreq(n_fft, d=1/sr)

    plt.figure(figsize=(12,3))
    plt.plot(freq, amp_db)
    plt.title(title)
    plt.xlabel("frequency (Hz)")
    plt.ylabel("amplitude (dB, relative)")
    plt.xscale("log")
    plt.grid(True)
    plt.show()

plot_spectrum(wave, sr, n_fft=2048)
```

### Short-Time Fourier Transform
The Short-Time Fourier Transform (STFT) decomposes a signal into overlapping time frames and computes a Fourier Transform independently for each frame. This allows us to track how frequency content evolves over time.

The STFT is defined as:
$$
X(m,k)=\sum^{N-1}_{n=0}x[n+mH]w[n]e^{−j2\pi kn/N}
$$

where:
- $w[n]$ is a window function (e.g., Hann or Hamming)
- $N$ is the window size (number of samples per frame)
- $H$ is the hop length (stride between consecutive frames)
- $m$ indexes time frames
- $k$ indexes frequency bins

Each frame corresponds to a short, approximately stationary segment of the signal.
Thus, a spectrum is a snapshot at one time region. A spectrogram stacks many spectra over time using the STFT. This is one of the most informative visualizations for audio and is the backbone of many audio ML pipelines.

```python
def plot_spectrum_linear(x, sr, n_fft=2048, start_sample=0, fmax=8000):
    x = np.asarray(x)
    seg = x[start_sample:start_sample+n_fft]
    if len(seg) < n_fft:
        seg = np.pad(seg, (0, n_fft-len(seg)))
    seg = seg * np.hanning(n_fft)
    dft = np.fft.rfft(seg)
    amp = np.abs(dft) + 1e-12
    amp_db = 20*np.log10(amp / np.max(amp))
    freq = np.fft.rfftfreq(n_fft, d=1/sr)

    m = freq <= fmax
    plt.figure(figsize=(12,3))
    plt.plot(freq[m], amp_db[m])
    plt.title(f"Spectrum (linear axis, up to {fmax} Hz)")
    plt.xlabel("frequency (Hz)")
    plt.ylabel("amplitude (dB, relative)")
    plt.grid(True)
    plt.show()

plot_spectrum_linear(wave, sr, n_fft=2048, fmax=8000)
```

```python
def stft_mag_db(x, sr, n_fft=1024, hop_length=256):
    x = np.asarray(x, dtype=np.float32)
    if TORCHAUDIO_OK:
        # torchaudio expects torch tensor
        xt = torch.tensor(x)
        spec = torch.stft(
            xt, n_fft=n_fft, hop_length=hop_length, win_length=n_fft,
            window=torch.hann_window(n_fft), return_complex=True
        )
        mag = spec.abs().numpy()
    elif SCIPY_OK:
        f, t, Zxx = scipy.signal.stft(x, fs=sr, window="hann", nperseg=n_fft, noverlap=n_fft-hop_length, boundary=None)
        mag = np.abs(Zxx)
        # scipy gives (freq, time)
        return 20*np.log10(mag + 1e-8), f, t
    else:
        raise RuntimeError("Need torchaudio or scipy for STFT.")
    # torch.stft gives (freq, time)
    f = np.fft.rfftfreq(n_fft, d=1/sr)
    t = np.arange(mag.shape[1]) * hop_length / sr
    return 20*np.log10(mag + 1e-8), f, t

S_db, f, t = stft_mag_db(wave, sr, n_fft=1024, hop_length=256)

plt.figure(figsize=(12,4))
plt.imshow(S_db, origin="lower", aspect="auto",
           extent=[t.min(), t.max(), f.min(), f.max()])
plt.colorbar(label="dB")
plt.title("Spectrogram magnitude (dB)")
plt.xlabel("time (s)")
plt.ylabel("frequency (Hz)")
plt.show()
```

### Windowing and the Time–Frequency Trade-off
The window function $w[n]$ controls how the signal is weighted inside each frame. Common choices (Hann, Hamming) reduce spectral leakage by smoothly tapering the signal at frame boundaries.
There is an inherent time–frequency resolution trade-off:
- Large $N$ → high frequency resolution, low time resolution
- Small $N$ → high time resolution, low frequency resolution

This trade-off is fundamental and cannot be avoided; it reflects the uncertainty principle for signals.
In practice:
- speech tasks often use window sizes of 20–25 ms
- hop lengths are typically 10 ms (overlapping frames)

### Where problems show up in spectrograms
Spectrograms make many issues visible quickly: clipped recordings, background hum, narrow-band noise, missing frequency bands, long silences, or strong differences in loudness across examples.

Later, when your model makes mistakes, you should inspect spectrograms of failed examples to understand whether the failure comes from data, preprocessing, or modeling.

## 5) Mel Scale and Log-Mel Spectrograms
While spectrograms provide a time–frequency representation of audio signals, they use a linear frequency axis that does not match human auditory perception. In practice, both biological hearing systems and effective machine learning models benefit from representations that emphasize perceptually meaningful frequency resolution. The Mel scale and log-Mel spectrograms address this mismatch by incorporating perceptual priors directly into the signal representation.

Most audio Transformers do not attend directly on raw waveforms. Instead, they such log-mel spectrograms, which are a perceptually motivated, compressed version of the spectrogram.
In practice, this reduces dimensionality and helps models learn robust representations, especially in small-scale projects.

### Human Perception of Frequency
Human perception of pitch is approximately linear at low frequencies and logarithmic at high frequencies. In other words, equal spacing in physical frequency (Hz) does not correspond to equal spacing in perceived pitch.
The Mel scale is an empirical mapping from physical frequency $f$ (in Hz) to a perceptual scale $m$ (in mels), defined as:
$$
m=2595 \ \text{log}_{10}​(1+f/700)
$$

where:
- $f$ is the frequency in Hertz
- $m$ is the perceived frequency in mels

This transformation compresses high-frequency regions while preserving resolution at low frequencies, where speech-relevant phenomena such as formants and harmonics are concentrated.

### Mel Filterbanks
Rather than working directly in the Mel domain, we approximate the Mel scale using a bank of overlapping triangular filters applied to the linear-frequency spectrogram.
Let:
- $S \in \mathbb{R}^{T \times F}$ be the magnitude spectrogram,
- $F$ the number of FFT frequency bins,
- $M$ the number of Mel filters.

We construct a Mel filterbank matrix:
$$
W_{\text{mel}}​ \in \mathbb{R}^{F×M}
$$

where each column corresponds to a triangular filter centered at a specific Mel frequency.
The Mel spectrogram is obtained by:
$$
S_\text{mel}​=S \cdot W_\text{mel​}
$$
resulting in:
$$
S_\text{mel} \in \mathbb{R}^{T\times M}
$$
This operation performs a weighted aggregation of nearby frequency bins, reducing dimensionality while preserving perceptually important spectral information.

```python
def mel_filterbank(sr, n_fft, n_mels=64, fmin=0.0, fmax=None):
    # Simple mel filterbank implementation (triangular filters).
    # This is not optimized; it's meant for pedagogy and works well enough for this notebook.
    if fmax is None:
        fmax = sr / 2

    def hz_to_mel(hz):
        return 2595.0 * np.log10(1.0 + hz / 700.0)

    def mel_to_hz(mel):
        return 700.0 * (10.0**(mel / 2595.0) - 1.0)

    mels = np.linspace(hz_to_mel(fmin), hz_to_mel(fmax), n_mels + 2)
    hz = mel_to_hz(mels)
    bins = np.floor((n_fft + 1) * hz / sr).astype(int)
    bins = np.clip(bins, 0, n_fft//2)

    fb = np.zeros((n_mels, n_fft//2 + 1), dtype=np.float32)
    for m in range(1, n_mels + 1):
        left, center, right = bins[m-1], bins[m], bins[m+1]
        if center == left:
            center += 1
        if right == center:
            right += 1
        for k in range(left, center):
            fb[m-1, k] = (k - left) / (center - left)
        for k in range(center, right):
            fb[m-1, k] = (right - k) / (right - center)
    return fb
```

### Log-Mel Spectrogram
Raw Mel energies exhibit a wide dynamic range and are poorly matched to both human loudness perception and neural network optimization. Humans perceive loudness approximately logarithmically, motivating a log-compression step.
We compute the log-Mel spectrogram as:
$$
X_\text{logmel​}=\text{log}(S_\text{mel}​+ \epsilon)
$$
where $\epsilon > 0$ is a small constant added for numerical stability.

This transformation:
- compresses large amplitude variations
- emphasizes relative spectral differences
- improves numerical conditioning during training

```python
def log_mel_spectrogram(x, sr, n_fft=1024, hop_length=256, n_mels=64):
    # magnitude spectrogram
    S_db, f, t = stft_mag_db(x, sr, n_fft=n_fft, hop_length=hop_length)
    # convert back from dB to linear magnitude (approx), then power
    S = 10**(S_db/20.0)
    P = S**2

    fb = mel_filterbank(sr, n_fft, n_mels=n_mels)
    mel_power = fb @ P  # (n_mels, time)
    log_mel = np.log(mel_power + 1e-8)
    return log_mel, t
```

```python
logmel, t2 = log_mel_spectrogram(wave, sr, n_fft=1024, hop_length=256, n_mels=64)

plt.figure(figsize=(12,4))
plt.imshow(logmel, origin="lower", aspect="auto",
           extent=[t2.min(), t2.max(), 0, logmel.shape[0]])
plt.colorbar(label="log power")
plt.title("Log-mel spectrogram")
plt.xlabel("time (s)")
plt.ylabel("mel bins")
plt.show()

print("log-mel shape:", logmel.shape)
```

### Why log-Mel features work well for Transformers
Log-Mel spectrograms provide several key advantages:
- Compact representation: fewer frequency bins than linear spectrograms.
- Perceptual alignment: frequencies important to human hearing are emphasized.
- Stable optimization: logarithmic scaling reduces gradient imbalance.
- Structured input: time frames naturally map to tokens for Transformers.

For these reasons, log-Mel spectrograms constitute the de facto standard input representation for modern Transformer-based audio models, including speech recognition, audio classification, and self-supervised audio pretraining.

## 6) The two biggest engineering issues in audio pipelines

The first issue is variable-length audio. Most datasets contain clips of different durations. Neural models need batching, which means you must either pad, crop, or use bucketing strategies.

The second issue is scale: loudness varies across recordings. Even after converting to floating point, you often need normalization (per-clip, per-dataset, or using a fixed reference) to make optimization stable.

In the next cell, you will build a minimal collate function that pads waveforms to a common length.


```python
def pad_collate(batch, target_sr=16000, max_duration_s=1.0):
    """
    batch: list of dicts with keys: "audio" (Audio(decode=False)) and "label"
    Returns:
      x: (B, L) float32 waveforms padded/cropped to L = target_sr * max_duration_s
      y: (B,) int64 labels
    """
    waveforms = []
    labels = []
    max_len = int(target_sr * max_duration_s)

    for ex in batch:
        # Decode from bytes/path (no torchcodec needed)
        wav, sr = decode_audio(ex["audio"], target_sr=target_sr)  # wav is float32 mono

        # Crop or pad to max_len
        if len(wav) >= max_len:
            wav = wav[:max_len]
        else:
            wav = np.pad(wav, (0, max_len - len(wav)))

        waveforms.append(wav)
        labels.append(ex["label"])

    x = torch.tensor(np.stack(waveforms), dtype=torch.float32)
    y = torch.tensor(labels, dtype=torch.long)
    return x, y

# Try collating a small batch
batch = [ds[i] for i in range(8)]
x, y = pad_collate(batch, target_sr=16000, max_duration_s=1.0)
print("x:", x.shape, "y:", y.shape, "x dtype:", x.dtype)
```

## 7) Visualizing audio augmentations

In supervised learning, augmentations improve robustness. In self-supervised learning, they define which variations should be considered the same example.

We visualize a few common augmentations in waveform space and in log-mel space: time shift, additive noise, and SpecAugment-style masking.


```python
def augment_time_shift(x, shift_samples):
    x = np.asarray(x)
    return np.roll(x, shift_samples)

def augment_add_noise(x, snr_db=15.0):
    x = np.asarray(x)
    sig_pow = np.mean(x**2) + 1e-12
    noise_pow = sig_pow / (10**(snr_db/10))
    noise = np.random.randn(*x.shape) * np.sqrt(noise_pow)
    return x + noise

def specaugment_mask(logmel, time_mask=12, freq_mask=8):
    M = logmel.copy()
    # time mask
    t0 = np.random.randint(0, max(1, M.shape[1]-time_mask))
    M[:, t0:t0+time_mask] = M.min()
    # freq mask
    f0 = np.random.randint(0, max(1, M.shape[0]-freq_mask))
    M[f0:f0+freq_mask, :] = M.min()
    return M

x0 = np.asarray(wave)
sr0 = sr

x_shift = augment_time_shift(x0, int(0.05*sr0))          # 50 ms shift
x_noise = augment_add_noise(x0, snr_db=10.0)              # stronger noise

plot_waveform(x0, sr0, title="Original waveform (cropped)")
plot_waveform(x_shift, sr0, title="Time-shifted waveform (+50ms, cropped)")
plot_waveform(x_noise, sr0, title="Noisy waveform (SNR=10 dB, cropped)")

lm0, _ = log_mel_spectrogram(x0, sr0, n_fft=1024, hop_length=256, n_mels=64)
lm_noise, _ = log_mel_spectrogram(x_noise, sr0, n_fft=1024, hop_length=256, n_mels=64)
lm_mask = specaugment_mask(lm0, time_mask=18, freq_mask=10)

def plot_logmel(lm, title):
    plt.figure(figsize=(12,4))
    plt.imshow(lm, origin="lower", aspect="auto")
    plt.colorbar(label="log power")
    plt.title(title)
    plt.xlabel("time frames")
    plt.ylabel("mel bins")
    plt.show()

plot_logmel(lm0, "Original log-mel")
plot_logmel(lm_noise, "Log-mel after additive noise")
plot_logmel(lm_mask, "Log-mel after SpecAugment-style masking")
```
